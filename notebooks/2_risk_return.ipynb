```python
df = pd.read_csv("../data/raw/portfolio_allocation.csv")

# stima semplice di contribution to return mensile
df["monthly_contrib"] = df["weight_pct"] * df["monthly_return"] / 100
df_asset = df.groupby("asset_class", as_index=False)[["monthly_contrib"]].sum()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(data=df_asset, x="asset_class", y="monthly_contrib")
plt.title("Monthly contribution to return by asset class")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

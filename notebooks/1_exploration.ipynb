```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/raw/portfolio_allocation.csv")
df["weight_pct"] = df["weight_pct"].astype(float)
```

```python
df_asset = df.groupby("asset_class", as_index=False)["weight_pct"].sum()

plt.figure(figsize=(6,4))
sns.barplot(data=df_asset, x="asset_class", y="weight_pct")
plt.title("Portfolio allocation by asset class")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

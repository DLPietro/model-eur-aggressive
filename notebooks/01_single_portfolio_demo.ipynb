```python
# 0. Import base
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")

# 1. Carica il file pulito (puoi usare direttamente il CSV allegato)
df = pd.read_csv("/model-eur-aggressive/data/raw/model_eur_aggressive_clean.csv")

# Colonne attese: ['asset_class','isin','security_name','allocation_pct', ...]
# Se i nomi sono diversi, mappa qui:
df.columns = [c.strip().lower() for c in df.columns]
# esempio di normalizzazione
df = df.rename(columns={
    "allocation": "allocation_pct",
    "weight": "allocation_pct",
    "asset_group": "asset_class"
})

# Normalizza pesi in 0–1
if df["allocation_pct"].max() > 1:
    df["allocation_w"] = df["allocation_pct"] / 100
else:
    df["allocation_w"] = df["allocation_pct"]

df = df[df["allocation_w"] > 0].copy()
df.head()
```

```python
df_asset = df.groupby("asset_class", as_index=False)["weight_pct"].sum()

plt.figure(figsize=(6,4))
sns.barplot(data=df_asset, x="asset_class", y="weight_pct")
plt.title("Portfolio allocation by asset class")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

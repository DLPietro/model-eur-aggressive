```python
# Model EUR Aggressive - Complete Analysis
# Data → Analysis → Plots → Recommendations

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
from typing import List, Dict, Optional, Tuple
from datetime import datetime
import warnings
warnings.filterwarnings("ignore")

plt.style.use("seaborn-v0_8-darkgrid")
sns.set_palette("husl")
TRADING_DAYS = 252
DEFAULT_RF = 0.02

print("✓ Portfolio Engine Ready")
print(f"Date: {datetime.now().strftime('%Y-%m-%d %H:%M')}")
```

```python
# Data Loading

def load_portfolio(path: str) -> pd.DataFrame:
    """Carica CSV: ISIN, Nome, Allocazione (%). Normalizza e aggiunge liquidità"""
    df = pd.read_csv(path)
    df.columns = df.columns.str.strip().str.lower()
    
    isin_col = next((c for c in df.columns if "isin" in c), None)
    name_col = next((c for c in df.columns if any(x in c for x in ["name", "nome", "security"])), None)
    weight_col = next((c for c in df.columns if any(x in c for x in ["alloc", "peso", "weight", "%"])), None)
    
    if not all([isin_col, name_col, weight_col]):
        raise ValueError(f"Colonne non trovate. Disponibili: {df.columns.tolist()}")
    
    df = df[[isin_col, name_col, weight_col]].copy()
    df.columns = ["isin", "name", "allocation_pct"]
    
    df["allocation_pct"] = pd.to_numeric(df["allocation_pct"].astype(str).str.replace("%", ""), errors="coerce")
    df = df[df["allocation_pct"] > 0].copy()
    df["isin"] = df["isin"].astype(str).str.strip().str.upper()
    df = df[df["isin"].str.len() >= 5]
    
    # Normalizza a 100%
    total = df["allocation_pct"].sum()
    df["allocation_w"] = df["allocation_pct"] / total
    
    return df[["isin", "name", "allocation_w"]].reset_index(drop=True)

def map_isin_to_ticker(isin: str, name: str) -> str:
    """Mapping ISIN → ticker Yahoo (da completare manualmente per i tuoi asset)"""
    mapping = {
        "LU0248184466": "0P00006Q25.L",  # Schroder ISF Asian Opportunities EUR A
        "FR0007054358": "MEUD.PA",        # Amundi EURO STOXX 50 II ETF
        "LU0297942434": "0P0000XMSJ.L",  # BGF Global Corporate Bond A2 EUR H
        "IE00BD5CTX77": "SHYD.L",         # BNY Mellon Global Short-Dated HY Bond
        "LU0140363002": "0P00009QZM.L",  # Franklin Mutual European Fund A EUR
        "LU0106817157": "0P00006Q2A.L",  # Schroder SISF Emerging Europe EUR A
        "IE00B3ZW0K18": "EQQQ.L",         # iShares S&P 500 EUR Hedged ETF
        "CASH-LIQ": "CASH"
    }
    return mapping.get(isin, f"UNKNOWN_{isin[:8]}")

def download_prices(tickers: List[str], start: str = "2020-01-01", end: Optional[str] = None) -> pd.DataFrame:
    """Download prezzi da yfinance"""
    valid_tickers = [t for t in tickers if t != "CASH"]
    if not valid_tickers:
        return pd.DataFrame()
    
    data = yf.download(valid_tickers, start=start, end=end, progress=False)["Adj Close"]
    if isinstance(data, pd.Series):
        data = data.to_frame(valid_tickers[0])
    
    # Aggiungi cash (flat line a 1.0)
    if "CASH" in tickers:
        data["CASH"] = 1.0
    
    return data.ffill().dropna(how="all")

print("✓ Data loading functions defined")
```

```python
# Risk Metrics

def annualized_return(rets: pd.Series) -> float:
    return rets.mean() * TRADING_DAYS

def annualized_volatility(rets: pd.Series) -> float:
    return rets.std() * np.sqrt(TRADING_DAYS)

def sharpe_ratio(rets: pd.Series, rf: float = DEFAULT_RF) -> float:
    excess = rets - rf / TRADING_DAYS
    return annualized_return(excess) / annualized_volatility(rets)

def sortino_ratio(rets: pd.Series, rf: float = DEFAULT_RF) -> float:
    excess = rets - rf / TRADING_DAYS
    downside = excess[excess < 0]
    if len(downside) == 0:
        return np.inf
    downside_vol = downside.std() * np.sqrt(TRADING_DAYS)
    return annualized_return(excess) / downside_vol

def max_drawdown(rets: pd.Series) -> float:
    cum = (1 + rets).cumprod()
    peak = cum.cummax()
    dd = (cum / peak) - 1
    return dd.min()

def calmar_ratio(rets: pd.Series) -> float:
    ann_ret = annualized_return(rets)
    mdd = max_drawdown(rets)
    return ann_ret / abs(mdd) if mdd != 0 else np.inf

def historical_var(rets: pd.Series, alpha: float = 0.95) -> float:
    return np.quantile(rets, 1 - alpha)

def compute_metrics(rets: pd.Series, name: str = "Portfolio") -> Dict:
    """Calcola tutte le metriche per una serie di rendimenti"""
    return {
        "name": name,
        "ann_return": annualized_return(rets),
        "ann_volatility": annualized_volatility(rets),
        "sharpe": sharpe_ratio(rets),
        "sortino": sortino_ratio(rets),
        "max_drawdown": max_drawdown(rets),
        "calmar": calmar_ratio(rets),
        "var_95": historical_var(rets, 0.95),
        "skew": rets.skew(),
        "kurtosis": rets.kurtosis()
    }

print("✓ Risk metrics functions defined")
```

```python
# Analytics and Visualization

def plot_allocation_pie(df: pd.DataFrame, title: str = "Portfolio Allocation"):
    fig, ax = plt.subplots(figsize=(9, 6))
    df.set_index("name")["allocation_w"].plot(kind="pie", autopct="%1.1f%%", ax=ax, startangle=90)
    ax.set_ylabel("")
    ax.set_title(title, fontsize=14, fontweight="bold")
    plt.tight_layout()
    plt.show()

def plot_correlation_heatmap(corr: pd.DataFrame, title: str = "Correlation Matrix"):
    fig, ax = plt.subplots(figsize=(10, 8))
    sns.heatmap(corr, annot=True, cmap="RdBu_r", center=0, fmt='.2f', 
                square=True, linewidths=0.5, cbar_kws={"shrink": 0.8}, ax=ax)
    ax.set_title(title, fontsize=14, fontweight="bold")
    plt.tight_layout()
    plt.show()

def plot_equity_curve(rets: pd.Series, bench_rets: Optional[pd.Series] = None, title: str = "Cumulative Returns"):
    fig, ax = plt.subplots(figsize=(12, 6))
    cum_port = (1 + rets).cumprod()
    cum_port.plot(ax=ax, label="Portfolio", linewidth=2)
    
    if bench_rets is not None:
        cum_bench = (1 + bench_rets).cumprod()
        cum_bench.plot(ax=ax, label="Benchmark", linewidth=2, linestyle="--")
    
    ax.set_title(title, fontsize=14, fontweight="bold")
    ax.set_xlabel("Date")
    ax.set_ylabel("Cumulative Return")
    ax.legend()
    ax.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

def plot_drawdown(rets: pd.Series, title: str = "Drawdown"):
    cum = (1 + rets).cumprod()
    peak = cum.cummax()
    dd = (cum / peak) - 1
    
    fig, ax = plt.subplots(figsize=(12, 5))
    dd.plot(ax=ax, color="red", linewidth=1.5)
    ax.fill_between(dd.index, dd.values, 0, color="red", alpha=0.3)
    ax.set_title(title, fontsize=14, fontweight="bold")
    ax.set_xlabel("Date")
    ax.set_ylabel("Drawdown (%)")
    ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.0%}'))
    ax.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

def print_metrics_table(metrics_list: List[Dict]):
    df = pd.DataFrame(metrics_list).T
    df.index.name = "Metric"
    print("\n" + "="*60)
    print("PERFORMANCE & RISK METRICS")
    print("="*60)
    print(df.to_string())/home/pietro/Scrivania/Repositories/model-eur-aggressive/data
    print("="*60 + "\n")

print("✓ Analytics & visualization functions defined")
```

```python
# Loading Portfolio

portfolio_path = "model-eur-aggressive/data/model_eur_aggressive_clean.csv"
df_portfolio = load_portfolio(portfolio_path)

print("✓ Portfolio ready")
print(f"\nTotal Assets: {len(df_portfolio)}")
print(f"Total Weight: {df_portfolio['allocation_w'].sum():.1%}\n")
display(df_portfolio)

# Adding Tickers
df_portfolio["ticker"] = df_portfolio.apply(lambda x: map_isin_to_ticker(x["isin"], x["name"]), axis=1)
print("\n✓ Ticker mappati")
display(df_portfolio[["name", "ticker", "allocation_w"]])
```
